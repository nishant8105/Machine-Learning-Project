```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/Teen_Mental_Health_Dataset.csv')
```

```python
df.head()
```

```python
df['usage_intensity'] = df['daily_social_media_hours'] / df['sleep_hours']
df.head()
```

```python
df['stress_per_hour'] = df['stress_level'] / df['daily_social_media_hours']
df
```

```python
df['anxiety_sleep_ratio'] = df['anxiety_level'] / df['sleep_hours']
```

```python
df['usage_category'] = pd.cut(
    df['daily_social_media_hours'],
    bins=[0,2,5,10],
    labels=['Low','Medium','High']
)

df['sleep_category'] = pd.cut(
    df['sleep_hours'],
    bins=[0,5,7,10],
    labels=['Poor','Average','Good']
)
```

```python
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.select_dtypes(include=['object', 'category']).columns:
    df[col] = le.fit_transform(df[col])
```

```python
df.sample(15)
```

```python
df =df.astype('float')
df.info()
```

```python
df.to_csv('../data/processed/process_Teen_mental_health.csv', index=False)
```
