```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/Teen_Mental_Health_Dataset.csv")
df.head()
```

```python
nums_col = df.select_dtypes(include = ['int64', 'float64']).columns
print(nums_col)

for col in nums_col:
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

```python
sns.boxplot(x='gender', y='stress_level', data=df)
plt.show()
sns.boxplot(x='platform_usage', y='addiction_level', data=df)
plt.show()
```
